```python
for col in columns:
    plt.figure()
    sns.boxplot(x=vitals[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

7. Correlation Analysis

```python
numeric_cols = vitals.select_dtypes(include=['number'])

plt.figure(figsize=(8,6))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

🟢 PART 2 — OPEN PATIENTS (TEXT DATA)

1Load JSON Dataset

```python
import json

with open("../data/raw/open_patients.jsonl", "r") as f:
    patients = [json.loads(line) for line in f]

patients_df = pd.DataFrame(patients)
patients_df.head()
```

2.Inspect Structure

```python
patients_df.info()
```

3.Text Length Analysis

```python
patients_df['text_length'] = patients_df['description'].apply(len)

sns.histplot(patients_df['text_length'])
plt.title("Text Length Distribution")
plt.show()
```

4.Most Common Words

```python
from collections import Counter
import re

def clean_text(text):
    return re.findall(r'\b\w+\b', text.lower())

all_words = []

for text in patients_df['description']:
    all_words.extend(clean_text(text))

common_words = Counter(all_words).most_common(30)
common_words
```

```python
words, counts = zip(*common_words)

plt.figure(figsize=(10,5))
sns.barplot(x=list(counts), y=list(words))
plt.title("Most Common Words")
plt.show()
```
